```python
%matplotlib inline

import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
session = pd.read_csv('Data/sessions.csv', index_col = 'Unnamed: 0')
final = pd.read_csv('Data/final_grades.csv',index_col = ['Unnamed: 0'])
inter = pd.read_csv('Data/inter_grades.csv')
```

```python
deeds = session[session.activity == " Deeds"]
```

```python
# convert start and end time column dtypes to datetime    
for col in ['start_time', 'end_time']:
    deeds[col] =  pd.to_datetime(deeds[col], format='%Y-%m-%d %H:%M:%S')

# calculate the time difference between end and start time in seconds
deeds['time_delta']=(pd.to_timedelta(deeds['end_time'])-pd.to_timedelta(deeds['start_time'])).astype('timedelta64[s]')
```

```python
ag = deeds.groupby(['student_id'],as_index=False).sum()
```

```python
merged = pd.merge(ag, final, how='left',on=['student_id'])
merged
```

```python
merged.corr()
```
